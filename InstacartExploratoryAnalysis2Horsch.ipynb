## Breaking Down by Order Size

```
# Setting up SQL Query to aggregate order sizes by day

conn = sqlite3.connect('aisles_database.db')
order_products_prior.to_sql('order_products_prior', conn)
order_products_query = '''
    SELECT order_id, COUNT(*) AS products_total
    FROM order_products_prior
    GROUP BY order_id;
    '''
```

```
# Transform back into dataframe and view 

order_sizes_df = pd.read_sql(order_products_query, conn)
order_sizes_df.head()
```

```
# Plotting histogram for number of products per order

plt.figure(figsize = (12, 10))
plt.hist(order_sizes_df.products_total, color='steelblue')
plt.axvline(order_sizes_df.products_total.mean(), color = 'lightgreen', linestyle = '--', label = 'Mean')
plt.axvline(order_sizes_df.products_total.median(), color = 'orange', linestyle = '--', label = 'Median')
plt.title('Number of Products per Order')
plt.xlabel('Number of Products')
plt.ylabel('Number of Orders')
plt.yticks([0, 0.5e6, 1.0e6, 1.5e6, 2.0e6, 2.5e6], ['0', '500k', '1M', '1.5M', '2M', '2.5M'])
sns.despine()
plt.legend()
plt.savefig('products_per_order_hist.png', dpi=300, pad_inches=0.5)
plt.show()
```

```
# Set up SQL Query to join orders and order_products_prior

orders.to_sql('orders', conn)
orders_change_byday_query = '''
    SELECT o.order_id, 
           o.order_dow,
           COUNT(op.product_id) AS products_total
    FROM orders AS o
    INNER JOIN order_products_prior AS op
    ON o.order_id = op.order_id
    GROUP BY o.order_dow, o.order_id;
    '''
```

### Look at Most Popular Departments

```
# Set up SQL Query

aisles.to_sql('aisles', conn)
departments.to_sql('departments', conn)
```

```
# Write SQL Query

aisles_departments_query = '''
    SELECT p.product_id,
           p.product_name,
           p.aisle_id,
           a.aisle,
           p.department_id,
           d.department,
           COUNT(op.product_id) AS count
    FROM products AS p
    LEFT JOIN order_products_prior AS op
    ON p.product_id = op.product_id
    LEFT JOIN departments AS d
    ON p.department_id = d.department_id
    LEFT JOIN aisles AS a
    ON p.aisle_id = a.aisle_id
    GROUP BY 1, 2, 3, 4, 5, 6;
    '''
```

```
# Transform back into dataframe and view

aisles_departments_df = pd.read_sql(aisles_departments_query, conn)
aisles_departments_df.head()
```

```
# Look at products ordered by department and sort descending order by most orders

orders_by_department = (pd.DataFrame(aisles_departments_df.groupby(['department_id', 'department'])['count'].sum()).reset_index())
orders_by_department.sort_values('count', ascending=False, inplace=True)
orders_by_department['department'] = [x.capitalize() for x in orders_by_department['department']]
orders_by_department.head()
```

```
# Visualize Total Orders by Most Popular Departments

plt.figure(figsize= (12, 10))
sns.barplot(x='count', y='department', data=orders_by_department, color='steelblue')
plt.title('Total Products Ordered by Department')
plt.xlabel('Total Products Ordered')
plt.ylabel('Department Name')
plt.xticks([0, 1e6, 2e6, 3e6, 4e6, 5e6, 6e6, 7e6, 8e6, 9e6], ['0', '1M', '2M', '3M', '4M', '5M', '6M', '7M', '8M', '9M'])
plt.tight_layout()
sns.despine()
plt.savefig('orders_most_popular_departments.png', dpi=300, pad_inches=0.5)
plt.show()
```

### Look at Most Popular Aisles

```
# Look at products ordered by aisle and sort descending order by most orders

orders_by_aisle = (pd.DataFrame(aisles_departments_df.groupby(['aisle_id', 'aisle'])['count'].sum()).reset_index())
orders_by_aisle.sort_values('count', ascending=False, inplace=True)
orders_by_aisle['aisle'] = [x.capitalize() for x in orders_by_aisle['aisle']]
orders_by_aisle.head()
```

```
# Visualize Total Orders by Top 25 Most Popular Aisles

plt.figure(figsize=(12, 10))
sns.barplot(x='count', y='aisle', data=orders_by_aisle[:25], color='steelblue')
plt.title('Total Products Ordered by Aisle')
plt.xlabel('Total Products Ordered')
plt.ylabel('Aisle Name')
plt.xticks([0, 0.5e6, 1e6, 1.5e6, 2e6, 2.5e6, 3e6, 3.5e6], ['0', '0.5M', '1M', '1.5M', '2M', '2.5M', '3M', '3.5M'])
plt.tight_layout()
sns.despine()
plt.savefig('orders_most_popular_aisles.png', dpi=300, pad_inches=0.5)
plt.show()
```
